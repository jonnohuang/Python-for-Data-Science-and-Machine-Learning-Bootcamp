```python
from pyspark import SparkContext
```

```python
sc = SparkContext()
```

```python
%%writefile example.txt
first line
second line
third line
fourth line
```

```python
# The following converts text file to a RDD (Resilient Distributed Datasets) instance.  RDD is a fundamental data structure of Spark)
```

```python
textFile = sc.textFile('example.txt')
```

```python
textFile.count()
```

```python
textFile.first()
```

```python
secfind = textFile.filter(lambda line: 'second' in line)
```

```python
secfind
```

```python
secfind.collect()
```
